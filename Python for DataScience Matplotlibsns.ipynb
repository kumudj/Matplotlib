### Line Graph

```python
x=np.array([1,2,6,8])
y=np.array([3,8,1,10])
plt.plot(x,y,'g-o',ms=20,mec='r')  #marker egde colour
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
plt.plot(x,y)
plt.grid()
```

### plotting subplots

```python
plt.subplot(1,4,1)
plt.plot(x,y,'b--')
plt.subplot(1,4,2)
plt.plot(x,y,'g-o')
plt.subplot(1,4,3)
plt.plot(x,y,'g-*')
plt.subplot(1,4,4)
plt.plot(x,y,'g--')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b--')
plt.subplot(2,2,2)
plt.plot(x,y,'g-o')
plt.subplot(2,2,3)
plt.plot(x,y,'g-*')
plt.subplot(2,2,4)
plt.plot(x,y,'g--')
plt.show()
```

```python
#Graph inside a graph
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])  #left bottom width height 
ax2=fig.add_axes([0.6,0.4,0.4,0.4])
ax1.plot(x,y)
ax2.plot(y,x)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(x,y,'g',label='x vs y')
ax1.plot(y,x,'b',label='y vs x')
ax1.legend()
```

## matplotlib bars

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,color="red",width=0.6)
plt.show()
```

### Pie graph

```python
Value=[20,70,10,50]
cat=["apples","banana","orange","pineapple"]
plt.pie(Value,labels=cat)
plt.legend()
plt.show()
```

```python
Value=[20,70,10,50]
cat=["apples","banana","orange","pineapple"]
Explode=[0.1,0,0,0.5]
plt.pie(Value,labels=cat,explode=Explode)
plt.legend()
plt.show()
```

```python
df=sns.load_dataset('tips')
```

```python
df.info()
```

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="day")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="sex",col="day",col_wrap=2)
```

```python
sns.distplot(df.tip,kde=True)
```

```python
sns.barplot(data=df,x="day",y="tip")
```

```python
sns.barplot(data=df,x="sex",y="tip",hue="day",ci=None)
```

```python
sns.countplot(data=df,x="sex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=df)
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```
